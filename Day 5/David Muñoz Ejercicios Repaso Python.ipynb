## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6, 8, 5, 2, 0, 6, 5]
contador = 1

for num in nums:
    if num == 0:
        break
    contador += 1

print("Contador:", contador)

#Respuesta:
#Debe dar 5
```

## Ejercicio 2

Sumar los valores que hay en una lista de enteros. Sumar solo hasta el número 0.

```python
nums = [6, 8, 5, 2, 0, 6, 5]
suma = 0

for num in nums:
    if num == 0:
        break
    suma += num

print("Suma:", suma)

#Respuesta:
#Debe dar 21
```
